```python
import pandas as pd
zipCodeData = pd.read_csv("zipcode_demog.csv")
zipCodeData
```

```python
zipCodeData.dropna(how = 'any', inplace= True)
pd.to_numeric(zipCodeData.POPULATION)
pd.to_numeric(zipCodeData.NETTRAFFIC)

type(zipCodeData.NETTRAFFIC)
```

```python
import seaborn as sns
from pylab import savefig
sns.set(style="whitegrid")
Population_chart = sns.barplot(x="STATION", y="POPULATION", data=zipCodeData, palette=("winter"),label = None)
Population_chart.set_xticklabels(zipCodeData.STATION,rotation= 90)
Population_chart.set_xlabel("")
Population_chart.set_ylabel('POPULATION',size = 20)
Population_chart2 =Population_chart.twinx()
zipCodeData.plot(x="STATION", y="NETTRAFFIC", ax=Population_chart2, legend=True, color="r",grid=False, linewidth = 5)

%config InlineBackend.figure_format = 'svg'

#sns.set_context("poster")
#sns.set_context("poster", font_scale= 1)
```
